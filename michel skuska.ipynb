```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
from PIL import Image

img = cv.imread('obr5.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

img= cv.resize(img, (200,100), interpolation = cv.INTER_AREA)
velkost= img.shape
img= img[0:velkost[0]//2+20]

cv.imwrite('img.jpg', img)

img = cv.blur(img, (4, 4))
cv.imwrite('img.jpg', img)

_, binary = cv.threshold(img, 220, 255, cv.THRESH_BINARY)

Canny = cv.Canny(img, 20, 200)
kernel = cv.getStructuringElement(cv.MORPH_RECT, (20, 20))
binary = cv.erode(binary, kernel)
binary =cv.bitwise_not(binary)
cv.imwrite('Canny.jpg', Canny)
cv.imwrite('binary.jpg', binary)
vys= cv.bitwise_and(binary, Canny)


cv.imwrite('vys.jpg', vys)

kernel = cv.getStructuringElement(cv.MORPH_RECT, (1, 6))
vys = cv.erode(vys, kernel)
cv.imwrite('vys2.jpg', vys)
```
